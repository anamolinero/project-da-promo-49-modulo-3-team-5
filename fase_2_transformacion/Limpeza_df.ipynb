11. Cambiar NaN por non_travel

```python
df["businesstravel"].unique()
```

```python
def cambiar_vacio(cadena):
    # Reemplaza cadenas vacías por non_travel.
    if isinstance(cadena, str) and cadena == " ":
        return "non_travel"
    # Reemplaza valores NaN por non_travel.
    elif pd.isna(cadena): # Verifica si el valor cadena es NaN. Si es NaN, devuelve True; de lo contrario, devuelve False.
        return "non_travel"
    elif cadena == "non-travel": #Cambiar la línea media
        return "non_travel"
    else:
        return cadena
```

```python
df["businesstravel"] = df["businesstravel"].apply(cambiar_vacio)

df["businesstravel"].unique()
```

12. - Quitar simbolo del "$"
    - Remplazar "," por "." 
    - Convertir a float

```python
df[["$_monthlyincome", "$_monthlyrate", "$_salary","performancerating", "totalworkingyears", "worklifebalance", "yearsincurrentrole"]]
```

```python
def limpiar_columna(serie):
    return (
        serie
        .astype(str)           # aseguro que todo sea string
        .str.replace("$", "")  # quito $
        .str.replace(",", ".")  # cambio comas
        .astype(float)         # convierto a número decimal
    )
```

```python
columnas_a_convertir = ["$_monthlyincome", "$_monthlyrate", "$_salary","performancerating", "totalworkingyears", "worklifebalance", "yearsincurrentrole"]

for col in columnas_a_convertir:
    df[col] = limpiar_columna(df[col])
```
