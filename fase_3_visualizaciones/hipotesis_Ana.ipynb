## Visualizaciones y gráficas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_final = pd.read_csv("fase_2_limpieza/df_final.csv",index_col=0) 
df_final.head()
```

Relación entre trabajo remoto y retención de empleados.

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df_final, x="remotework", hue="attrition", palette="pastel")
plt.xlabel("Trabajo Remoto")
plt.ylabel("Cantidad de Empleados")
plt.title("Relación entre Trabajo Remoto y Retención de Empleados")
plt.legend(title="Attrition", labels=["No (Retenido)", "Yes (Renunció)"])
plt.show()
```

```python
# Distribución de salarios según trabajo remoto
sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(data=df_final, x="remotework", y="$_salary", palette="pastel", ax=axes[0])
axes[0].set_title("Distribución de Salarios según Trabajo Remoto")
axes[0].set_xlabel("Trabajo Remoto")
axes[0].set_ylabel("Salario")

plt.tight_layout()
plt.show()
```

```python
# Comparación de satisfacción laboral con balance vida-trabajo
sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(data=df_final, x="attrition", y="worklifebalance", palette="magma", ax=axes[1])
axes[1].set_title("Satisfacción laboral vs. Balance vida-trabajo")
axes[1].set_xlabel("Attrition (renunció)")
axes[1].set_ylabel("Balance vida-trabajo")

plt.tight_layout()
plt.show()
```

```python
# Relación entre años en la empresa y la renuncia
sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(data=df_final, x="attrition", y="yearsatcompany", palette="pastel", ax=axes[2])
axes[2].set_title("Años en la empresa vs. renuncia")
axes[2].set_xlabel("Attrition (renunció)")
axes[2].set_ylabel("Años en la empresa")

plt.tight_layout()
plt.show()
```

```python
#Distribución de edades
sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(df_final["age"], bins=20, kde=True, color="blue")
plt.title("Distribución de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Distribución de salarios
plt.figure(figsize=(8, 5))
sns.histplot(df_final["$_salary"], bins=20, kde=True, color="green")
plt.title("Distribución de Salarios")
plt.xlabel("Salario")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Distribución de años de experiencia
plt.figure(figsize=(8, 5))
sns.histplot(df_final["totalworkingyears"], bins=20, kde=True, color="purple")
plt.title("Distribución de Años de Experiencia")
plt.xlabel("Años de Experiencia")
plt.ylabel("Frecuencia")
plt.show()
```

# Análisis de Resultados

- Attrition: La mayoría de los empleados no han dejado la compañía, pero hay un porcentaje significativo que sí.

- Satisfacción Laboral: Los empleados que se van tienden a reportar menores niveles de satisfacción laboral.

- Edad: Los empleados más jóvenes parecen tener mayor probabilidad de dejar la compañía.

- Ingreso Mensual: Los empleados que se van tienden a tener ingresos mensuales más bajos.

- Departamento: Ciertos departamentos muestran mayores tasas de attrition que otros.

- Balance Vida-Trabajo: Los empleados con menor balance vida-trabajo tienen mayor probabilidad de irse.

- Tiempo en la Compañía: Los empleados que llevan menos tiempo en la compañía tienen mayor probabilidad de irse.

- Trabajo Remoto: El trabajo remoto parece tener algún efecto en la retención de empleados.

- Satisfacción Ambiental: Los empleados menos satisfechos con su ambiente de trabajo tienen mayor probabilidad de irse.
